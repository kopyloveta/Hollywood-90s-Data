```python
import pandas as pd
import csv
import os

from dvt import video_info
from dvt import AnnoShotBreaks
```

```python
anno_sb = AnnoShotBreaks()
```

```python
path_info = 'files_info'
path_shots = 'shot_detection' 

if not os.path.exists(path_info):
    os.makedirs(path_info)

if not os.path.exists(path_shots):
    os.makedirs(path_shots)
```

```python
files_data = pd.read_csv('Close-upUSA1990.tsv', sep='\t', header=0)
```

```python
for index, row in files_data.iterrows():
    print(f"file {index}")
    path = row['path']
    title = row['eng title'].replace(":", "").replace(r"\/", "")
    filename_info = f"{path_info}/info_{title}.csv"
    filename_shot = f"{path_shots}/shot-data_{title}.csv"
    with open(filename_info,'w') as f:
        w = csv.writer(f)
        w.writerows(video_info(path).items())
    out_sb = anno_sb.run(path)
    out_scene = pd.DataFrame(out_sb['scenes'])
    out_scene.to_csv(filename_shot, index=False)
```
