```python
import os
import re
import time
import datetime
```

```python
for index, row in df.iterrows():
    path = row["path"]
    if not path:
        continue
    if path.startswith('#'):
        path = path[1:]
        if path.endswith('.part'):
            path = path.replace('.part', '')
        if not os.path.exists(path):
            continue
    
    if re.search('[a-z] /home', path):
        #path2 = path.split(' /home')
        #path2[1] = '{} home/'.format(path2[1])
        continue
    
    vidcap = cv2.VideoCapture(path)
    success,image = vidcap.read()
    count = 0
    
    year = row["year"]
    genre = row["genre"]
    eng_title = row["eng title"]
    
    eng_title = eng_title.replace('/', ' ')
    eng_title = eng_title.replace("'", ' ')
    eng_title = eng_title.replace(' ', '_')
    
    path_dir = os.path.join('/media/boris/HD-E1/USA1990/frames/USA1990', str(year), genre, eng_title) 
    if not os.path.exists(path_dir):
        os.makedirs(path_dir)
    else:
        continue
    print(path)
    df_fr.loc[index, 'path'] = path
    df_fr.loc[index, 'year'] = year
    df_fr.loc[index, 'genre'] = genre
    df_fr.loc[index, 'rus title'] = row["rus title"]
    df_fr.loc[index, 'eng title'] = row["eng title"]
    df_fr.loc[index, 'imdb'] = row["imdb"]
    df_fr.loc[index, 'mark'] = row["mark"]
    df_fr.loc[index, 'frames'] = path_dir
    
    while success:
        if count % 24 == 0:
            sec_frame = datetime.datetime(1,1,1) + datetime.timedelta(seconds=count/24)
            cv2.imwrite("{}/{}_{}-{}-{}.jpg".format(path_dir, count, sec_frame.hour, sec_frame.minute, sec_frame.second), image)
        success,image = vidcap.read()
        if not success:
            break
        count += 1
            
    df_fr.loc[index, 'count_frames'] = count
```

```python
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    #rolmean = pd.rolling_mean(timeseries, window=12)
    rolmean = timeseries.rolling(60).mean()
    #rolstd = pd.rolling_std(timeseries, window=12)
    rolstd = timeseries.rolling(60).std()

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    plt.title('Rolling Mean')
    plt.show(block=False)
              
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.title('Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
```

```python
def test_stationarity_pack(timeseries, pthtsv, ttl):
    
    #Determing rolling statistics
    #rolmean = pd.rolling_mean(timeseries, window=12)
    rolmean = timeseries.rolling(60).mean()
    #rolstd = pd.rolling_std(timeseries, window=12)
    #rolstd = timeseries.rolling(60).std()
    
    mean = plt.plot(rolmean, color='black', label='Rolling Mean')
    plt.title('{}: Rolling Mean'.format(ttl.strip()))
    plt.savefig('plots/{}-mean-comp.png'.format(pthtsv))
    #plt.show(block=False)
    plt.close()
    
    mean = plt.plot(rolmean, color='black', label='Rolling Mean')
    plt.title('{}: Rolling Mean'.format(ttl.strip()))
    plt.savefig('plots/{}-mean.png'.format(pthtsv))
    #plt.show(block=False)
    plt.close()
    
    
    #Perform Dickey-Fuller test:
    #print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    return dfoutput, rolmean
```
